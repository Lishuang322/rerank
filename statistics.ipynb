```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
import os
#compute the percentage of privacy region, and store them to a dictionary
dataset = "/data/data-hulishuang/img-dataset/oxford_smeared/white/t155_p400/mask"
percentage_dic = {}
more_than_half = 0
total = 0
for f in os.listdir(dataset):
    file_name = os.path.join(dataset,f)
    mask = io.imread(file_name,as_gray = True)
    
    white = np.count_nonzero(mask == 255)
    black = np.count_nonzero(mask == 0)
    percentage = np.true_divide(white,white+black)
    percentage = np.around(percentage,decimals = 2)
    if percentage == 0:
        print(f + str(white)+' ' + str(black))
    if percentage > 0.5:
        more_than_half = more_than_half + 1
    percentage_dic[f[:-4]] =  percentage
    total = total + 1
    #print(percentage_dic)
print(more_than_half)
print(total)
print('percentage of more than half:',more_than_half/total)    
#print(percentage_dic)    

#get query image for every rnk

#get percentage of query image from percentage_dic
```

```python

label = "/data/data-hulishuang/img-dataset/oxford_label"
percentage_query = {}
for fil in os.listdir(label):
    if fil[-9:]=="query.txt":
        
        with open(os.path.join(label,fil),'r') as f:
            data_q = f.readline().split(' ')
            query_img = data_q[0][5:]
            percentage_query[fil[:-10]] = percentage_dic[query_img]
            
print(percentage_query)  
```
